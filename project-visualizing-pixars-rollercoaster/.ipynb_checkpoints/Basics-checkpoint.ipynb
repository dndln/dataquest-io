```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # this overrides the matplotlib styling
# jupyter magic to display plots inline
%matplotlib inline

pixar_movies = pd.read_csv('PixarMovies.csv')
num_rows = pixar_movies.shape[0]
num_columns = pixar_movies.shape[1]

print('Data:')
pixar_movies.head(num_rows)
```

```python
print('Data types:')
pixar_movies.dtypes
```

```python
print('Summary statistics:')
pixar_movies.describe()
# can't comput percentils of oscars won, because Inside Out is NaN
```

```python
pixar_movies.loc[:, 'Domestic %'] = pixar_movies.loc[:,'Domestic %'].str.strip('%')
pixar_movies.loc[:, 'International %'] = pixar_movies.loc[:, 'International %'].str.strip('%')
# print(pixar_movies.head(num_rows))

# .astype() takes both 'float' and float
pixar_movies.loc[:, 'Domestic %'] = pixar_movies.loc[:, 'Domestic %'].astype(float)
pixar_movies.loc[:, 'International %'] = pixar_movies.loc[:, 'International %'].astype('float')
pixar_movies.dtypes

```

```python
pixar_movies.loc[:, 'IMDB Score'] = pixar_movies.loc[:, 'IMDB Score'] * 10
print(pixar_movies['IMDB Score'])
```

```python
filtered_pixar = pixar_movies.dropna()
filtered_pixar
```

```python
pixar_movies.set_index('Movie', inplace=True)
pixar_movies
```

```python
filtered_pixar.set_index('Movie', inplace=True)
filtered_pixar
```

```python
critics_reviews = pixar_movies.loc[:, ['RT Score', 'IMDB Score', 'Metacritic Score']]
critics_reviews.plot(figsize=(10, 6))
plt.show()
```

```python
critics_reviews.plot(kind='box', figsize=(9, 5))
# this misses the outliers in RT score (lowest score is 39)
# There is an issue with matplotlib boxplot fliers not 
# showing up when seaborn is imported, even when 
# fliers are explicitly enabled. 
plt.show()
print("The matplotlib boxplot is missing the RT Score outlier at 39.")

plt.figure(figsize=(9,5))
sns.boxplot(data=critics_reviews, width=0.25)

sns.plt.show()

print(critics_reviews)
```
