```python
# Turn off chained assignment warning for recent_grads
recent_grads.is_copy = False

recent_grads.loc[:, 'ShareMen'] = recent_grads.loc[:, 'Men'] / recent_grads.loc[:, 'Total']

arts_vector = (recent_grads.loc[:, 'Major_category'] == 'Arts')
arts_rows = recent_grads.loc[arts_vector, :]

arts_rows.set_index('Major', inplace=True)
arts_rows.loc[:, ['ShareMen', 'ShareWomen']].plot(kind='bar')

# dataquest leaves plt.show() off their solutions
plt.show()
```

```python
# Grouped Bar Plot, Part 1

fig2 = plt.figure(figsize=(8, 8))
ax1_fig2 = fig2.add_subplot(1, 1, 1)

import numpy as np
locs = np.arange(len(arts_rows.index))

bar_1_width = 0.35
# this was a list of values, which was why the xticklabels were wrong
# arts_majors_list = arts_rows.loc[:, 'ShareMen'].tolist()
arts_majors_values = arts_rows.loc[:, 'ShareMen'].tolist()
arts_majors_list = arts_rows.index.tolist()
bar_1 = ax1_fig2.bar(left=locs, height=arts_majors_values, 
    width=bar_1_width)

# set_xticklabels is a method of a "figure.axes", created by ".add_subplot()"
ax1_fig2.set_xticklabels(arts_majors_list, rotation=90)

plt.show()
```

```python
# Grouped Bar Plot, Part 1

fig2 = plt.figure(figsize=(8, 8))
ax1_fig2 = fig2.add_subplot(1, 1, 1)

import numpy as np
locs = np.arange(len(arts_rows.index))

bar_1_width = 0.35
# this was a list of values, which was why the xticklabels were wrong
# arts_majors_list = arts_rows.loc[:, 'ShareMen'].tolist()
male_arts_majors_values = arts_rows.loc[:, 'ShareMen'].tolist()
arts_majors_list = arts_rows.index.tolist()
bar_1 = ax1_fig2.bar(left=locs, height=male_arts_majors_values, 
    width=bar_1_width)

# set_xticklabels is a method of a "figure.axes", created by ".add_subplot()"
ax1_fig2.set_xticklabels(arts_majors_list, rotation=90)

# plt.show()

# Part 2

# you can add elementwise to lists with "+", no need to loop
offset_locs = locs + bar_1_width
female_arts_majors_values = arts_rows.loc[:, 'ShareWomen'].tolist()
bar_2 = ax1_fig2.bar(left=offset_locs, 
    height=female_arts_majors_values, width=bar_1_width, color='green')
ax1_fig2.set_xticks(offset_locs)

# these two lines do the same thing in a different way
# first is called from plt.figure.axes
# second is called from plt
ax1_fig2.legend(['ShareMen', 'ShareWomen'], loc='upper left')
# plt.legend((bar_1, bar_2), ('ShareMen', 'ShareWomen'), loc='upper left')

# these two lines do the same thing, one is called from plt.figure.axes
# second is called from plt
ax1_fig2.grid()
# plt.grid()

fig2.suptitle('Male to Female Ratio by Major')

plt.show()
```

```python
# Pandas stacked boxplot

arts_m_f_ratio = arts_rows.loc[:, ['ShareMen', 'ShareWomen']]

fig3 = arts_m_f_ratio.plot.bar(stacked=True)
fig3.set_title('Male to Female Ratio by Major')
plt.show()
```
